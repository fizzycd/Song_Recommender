# What happens if the user inputs a song that doesn't exist?

```python
def ask_input_cat(question, options):
    while True:
        prompt = input(f'enter {question}: ')
        if (prompt not in options):
            print(f'Your choice is not in the Top100')
        else:
            return prompt
```
